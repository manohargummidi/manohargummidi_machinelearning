```python
##import required packages and models 
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#import dataset 
dataset = pd.read_csv('C:\\Users\\UCiK\\OneDrive\\Desktop\\ds course\\datasets\\Churn_Modelling.csv')
```

```python
#first 5 rows of dataset
dataset.head()
```

```python
#split the data into input and outputs
x= dataset.iloc[:,3:13]
y=dataset.iloc[:,13]
```

```python

#convert gender catagorical variable into numarical format
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['Gender']=le.fit_transform(x['Gender'])
```

```python
x
```

```python
#get dummyvariables in x dataset 
x=pd.get_dummies(x)
```

```python
#scale the data (normalization)

from sklearn.preprocessing import MinMaxScaler
nr = MinMaxScaler()
x_norm=nr.fit_transform(x)
x_norm
```

```python
#column names stored in the names 
names= x.columns
names
```

```python
#normalized data of x stored in x_n with add the column names 
x_n=pd.DataFrame(x_norm , columns=names )
```

```python
x_n
```

```python
#split the data into training and the testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_n , y , train_size= .75 , random_state=0)
```

```python
#model creating 

#logistic
from sklearn.linear_model import LogisticRegression 
cl = LogisticRegression()
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('LogisticRegression: ', score)

#################################################################################################################################
##support vector machine 

from sklearn.svm import SVC
cl = SVC(kernel = 'linear')
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('SVC linear: ', score)



from sklearn.svm import SVC
cl = SVC(kernel = 'sigmoid')
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('SVC sigmoid: ', score)

#################################################################################################################################

#Decision tree

from sklearn.tree import DecisionTreeClassifier
cl = DecisionTreeClassifier(criterion= 'entropy')
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('DecisionTreeClassifier: ', score)

#################################################################################################################################

##randomforestclassifier with 10 estimaters 

from sklearn.ensemble import RandomForestClassifier
cl = RandomForestClassifier(n_estimators=10 , criterion= 'entropy')
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('RandomForestClassifier n=10: ', score)

#################################################################################################################################

###randomforestclassifier with 150 estimaters 

from sklearn.ensemble import RandomForestClassifier
cl = RandomForestClassifier(n_estimators=150 , criterion= 'entropy')
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('RandomForestClassifier , n=150: ', score)


#################################################################################################################################

#gradientdecentboostingclassifier
from sklearn.ensemble import GradientBoostingClassifier
cl = GradientBoostingClassifier()
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('GradientBoostingClassifier: ', score)


#################################################################################################################################
 #naviebayesclassifier
    
from sklearn.naive_bayes import GaussianNB
cl = GaussianNB()
cl.fit(x_train,y_train)
y_pred= cl.predict(x_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
print('GaussianNB: ', score)
```
