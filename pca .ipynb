```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uni =pd.read_csv('C:\\Users\\UCiK\\OneDrive\\Desktop\\ds course\\datasets\\Universities.csv')
```

```python
del uni['Univ']
```

```python
uni
```

```python
dt = uni-uni.min()/(uni.max()-uni.min())
```

```python
dt
```

```python
uni.mean()
```

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components=4)
```

```python
pca_val = pca.fit_transform(dt)
```

```python
pca_val
```

```python
score=pd.DataFrame(pca_val)
```

```python
score
```

```python
var = pca.explained_variance_ratio_
```

```python
var
```

```python
var1 = np.cumsum(np.round(var , decimals=2)*100)
```

```python
var1
```

```python
princ_df = pd.DataFrame(data = pca_val , columns = ['pc1','pc2','pc3','pc4'])
```

```python
princ_df
```

```python
import scipy.cluster.hierarchy as scs
```

```python
dendogram =scs.dendrogram(scs.linkage(princ_df , method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clus = AgglomerativeClustering(n_clusters=2 , affinity="Euclidean" , linkage= 'single')
```

```python
abc = clus.fit_predict(princ_df)
```

```python
abc
```

```python
uni['groups']=pd.Series(abc)
```
